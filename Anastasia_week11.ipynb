```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import *
# reading image
img = cv2.imread("rainbow.jpg")
#print(img.shape)
img = cv2.resize(img, (200,150))
# display image using openCV
cv2_imshow(img)
color_edges = cv2.Canny(img, threshold1=400,threshold2=200)
cv2_imshow(color_edges)
print(color_edges.shape)

#edge extraction - convert to grayscale, blur the image,then extract the edges

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

gray = cv2.medianBlur(gray, 3) #smoothing it out
cv2_imshow(gray)

edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 15) #detecting the edges

cv2_imshow(edges)
```
